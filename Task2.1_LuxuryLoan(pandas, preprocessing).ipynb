```python
# data contingency
LuxuryLoanEdit = data.copy()
```

```python
# its column names
LuxuryLoanEdit.columns
```

```python
# dropping several columns for convenience
LuxuryLoanEdit= LuxuryLoanEdit.drop(['interest rate','duration months','firstname', 'middlename', 'lastname', 'social', 'phone', 'title',
       'employment length', 'BUILDING CLASS CATEGORY', 'TAX CLASS AT PRESENT',
       'BUILDING CLASS AT PRESENT', 'ADDRESS 1', 'ADDRESS 2', 'ZIP CODE',
       'CITY', 'STATE', 'TOTAL UNITS', 'LAND SQUARE FEET', 'GROSS SQUARE FEET',
       'TAX CLASS AT TIME OF SALE'],axis=1)
```

```python
# check new head
LuxuryLoanEdit.head()
```

```python
# renaming a column & final column checking
LuxuryLoanEdit = LuxuryLoanEdit.rename({'10 yr treasury index date funded': 'treasury index'}, axis=1) 
LuxuryLoanEdit.columns
```

```python
# a bit data cleaning
def change(cols):
    if cols == 'commerical property':
        cols = 'commercial property'
    return cols

LuxuryLoanEdit['purpose'] = LuxuryLoanEdit['purpose'].apply(change)
```

```python
# its checking
LuxuryLoanEdit['purpose'].value_counts()
```

```python
# saving to csv
LuxuryLoanEdit.to_csv('LuxuryLoanEdit.csv', index= False)
```
